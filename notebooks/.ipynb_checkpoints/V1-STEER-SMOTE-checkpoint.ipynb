# SMOTE or Synthetic Minority Oversampling Technique

**Here, the dataset has severe class imbalance**

>SMOTE works by selecting examples that are close in the feature space, drawing a line between the examples in the feature space and drawing a new sample at a point along that line.

# SMOTE for Balancing Data

```python
from numpy import where
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_classification
# importing SMOTE
from imblearn.over_sampling import SMOTE
```

```python
# define dataset
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
```

```python
# summarize class distribution
counter = Counter(y)
print(counter)
```

```python
# scatter plot of examples by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# transforming the data using SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
# summarize the new class distribution
counter = Counter(y)
print(counter)
```

```python
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.grid(True)
plt.show()
```

#  SMOTE - Extensions
> Borderline - SMOTE : involves selecting instances/attributres of minority class that are misclassified, such as with k-nearest neighbour classification model.

> Borderline - SMOTE SVM - instead of kNN, SVM is used to identify misclassified examples on the decision boundary.

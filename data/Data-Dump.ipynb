```python
import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime
```

```python
nr_of_customers = 10000

fake = Faker('de_DE')

customers = []

for customers_id in range(nr_of_customers):

    #Type of device
    d_type = random.choice(["Laptop", "Smart Phone", "Feature Phone"])
    
    #Device ID
    device_ID = fake.ean(length=8)
    
    #Device Brand
    if d_type == "Laptop":
        d_brand = random.choice(["Apple", "HP", "Dell", "Lenovo"])
    elif d_type == "Smart Phone":
        d_brand = random.choice(["Apple", "Blackberry", "OnePlus", "Oppo", "Nokia", "Samsung"])
    else:
        #feature phone
        d_brand = random.choice(["Nokia", "Samsung"])
    
    #Device model
    if d_brand == "Apple" and d_type == "Laptop":
        d_model = random.choice(["MacBook", "ProBook"])
    elif d_brand == "Lenovo":    
        d_model = random.choice(["Ideapad", "ThinkPad"])
    elif d_brand == "Dell":
        d_model = random.choice(["Inspiron", "Vostro"])
    elif d_brand == "HP":
        d_model = random.choice(["EliteBook", "Probook", "Pavilion"])
    elif d_brand == "Apple" and d_type == "Smart Phone":
        d_model = random.choice(["Iphone 5", "Iphone 5S", "Iphone SE", "Iphone 6", "Iphone 6S", "Iphone 7", "Iphone 8", "Iphone X", "Iphone XR", "Iphone XS", "Iphone 11", "Iphone 11", "Iphone 12", "Iphone 13"])
    elif d_brand == "Blackberry":
        d_model = random.choice(["Q5", "Q10", "Curve", "Torch", "Leap", "Passport", "Keyone", "Aurora"])
    elif d_brand == "OnePlus":
        d_model = random.choice(["OnePlus 6", "OnePlus 6T", "OnePlus 7", "OnePlus 7pro", "OnePlus 7T", "OnePlus nord", "OnePlus 8", "OnePlus 9R"])
    elif d_brand == "Oppo":
        d_model = random.choice(["A3s", "A52", "A11k", "Find 7", "N1", "Neo", "Joy", "R1"])
    elif d_brand == "Nokia" and d_type == "Smart Phone":
        d_model = random.choice(["X100", "G300", "T20", "G50", "XR20", "X10", "C20 Plus", "X20"])
    elif d_brand == "Samsung" and d_type == "Smart Phone":
        d_model = random.choice(["Galaxy A6", "Galaxy J7 duo", "Galaxy S9", "Galaxy Note 9", "Galaxy A8s"])
    elif d_brand == "Nokia" and d_type == "Feature Phone":
        d_model = random.choice(["5310", "125", "225", "215", "6300", "8000"])
    else:
        #samsung feature phone
        d_model = random.choice(["5310", "125", "225", "215", "6300", "8000"])
    
    
    #Manufacture date
    Md1 = datetime.strptime(f'01/01/2014', '%m/%d/%Y')
    Md2 = datetime.strptime(f'12/31/2020', '%m/%d/%Y')
    manufacture_date = fake.date_between(Md1, Md2)
    
    #Date of Allocation
    d1 = datetime.strptime(f'01/01/2021', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2021', '%m/%d/%Y')
    allocation_date = fake.date_between(d1, d2)
    
    #Date of Assessment
    ad1 = datetime.strptime(f'01/01/2022', '%m/%d/%Y')
    ad2 = datetime.strptime(f'02/22/2022', '%m/%d/%Y')
    assessment_date = fake.date_between(ad1, ad2)
    #datetime.strptime(f'02/18/2022', '%m/%d/%Y')
    
    
    #Age
    age = abs(assessment_date - manufacture_date).days
    
    #Water Damage
    water_damage = random.choice(["Yes", "No"])
    
    #Keypad, keyboard, touch responsive
    d_responsive = random.choice(["Yes", "No"])
    
    #Device in use
    if d_responsive == "No":
        device_in_use = "No"
    else:
        device_in_use = random.choice(["Yes", "No"])
    
    
    
    #Battery Health
    battery_health = random.choice(["Good", "Average", "Poor"])
    
    #usage(hrs)
    if battery_health == "Good":
        usage = fake.pyfloat(right_digits=2, positive=True, min_value=12, max_value=20)
        latency = "Good"
    elif battery_health == "Average":
        usage = fake.pyfloat(right_digits=2, positive=True, min_value=6, max_value=11)
        latency = "Average"
    else:
        usage = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=5)
        latency = "Poor"
    
    #Latency(ms)
    #latency = random.choice(["Good", "Average", "Poor"])
    
    customers.append([d_type, device_ID, d_brand, d_model, manufacture_date, allocation_date, assessment_date, age, device_in_use, water_damage, d_responsive, battery_health, usage, latency])

customers_df = pd.DataFrame(customers, columns=['Device_type','Device_ID', 'Device_brand', 'Device_model', 'Manufacture_Date', 'Allocation_Date', 'Assessment_Date', 'Age(days)','device_in_use', 'water_damage', 'device_responsive', 'battery_health', 'Usage(hrs)', 'Latency(ms)']) 
   
customers_df.to_csv('V5_Dummy-Data.csv')
pd.pandas.set_option('display.max_columns', None)
print(customers_df.head(10))
```
